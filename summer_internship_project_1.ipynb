## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing database


```python
df=pd.read_csv('countries-aggregated_till_10_july.csv')
df
```

## Daily Cases Calculation

```python
df['Daily_Confirmed'] = df['Confirmed'] - df['Confirmed'].shift(1)
df
```

## Spread Rate Calculation

```python
df['Spread Rate'] = round(((df['Daily_Confirmed'].shift(1)-df['Daily_Confirmed'])/df['Daily_Confirmed'])*100,2);
df
```

```python
ACTIVE_INDIA = (INDIA["Confirmed"] - INDIA["Recovered"] - INDIA["Deaths"]).values
plt.scatter(date, ACTIVE_INDIA, label="ACTIVE CASES")
plt.title("ACTIVE Cases in India")
plt.xlabel("Date")
plt.ylabel("ACTIVE ")
plt.legend()
plt.xticks(rotation=45)
plt.gcf().set_size_inches((15, 7))
```

```python
SPREAD_INDIA = INDIA["Spread Rate"].values
plt.scatter(date,SPREAD_INDIA, label="SPREAD RATE")
plt.title("Spread Rate in India")
plt.xlabel("Date")
plt.ylabel("Spread ")
plt.legend()
plt.xticks(rotation=45)
plt.gcf().set_size_inches((10, 7))
```

```python
yci1 = confirmed_INDIA
yci1 = yci1.values.reshape(-1,1)
xci1 = np.arange(len(date))
xci1 = xci1.reshape(-1,1)
```

```python

from sklearn.linear_model import LinearRegression
rci1 = LinearRegression()
rci1.fit(xci1,yci1)
        
ypci1 = rci1.predict(xci1)
        

from sklearn.preprocessing import PolynomialFeatures
polyci1 = PolynomialFeatures(degree = 2)
Xci1 = polyci1.fit_transform(xci1)

from sklearn.linear_model import LinearRegression
reci1 = LinearRegression()
reci1.fit(Xci1,yci1)

YPci1 = reci1.predict(Xci1)



from sklearn.preprocessing import StandardScaler
sxci1 = StandardScaler()
syci1 = StandardScaler()
SXci1 =  sxci1.fit_transform(xci1)
SYci1 =  syci1.fit_transform(yci1)

from sklearn.svm import SVR
regci1 = SVR(kernel = 'rbf')
regci1.fit(SXci1,SYci1.ravel())
SVYci1 = regci1.predict(SXci1)

# plt.plot(x, CONFIRMED, label=" Actual cases")
plt.scatter(sxci1.fit_transform(xci1), syci1.fit_transform(yci1),linewidth = 4, label=" Actual cases")
plt.plot(sxci1.fit_transform(xci1),syci1.fit_transform(ypci1), 'r',linewidth = 4,label = "Linear Regression : Accuracy = {}".format(rci1.score(xci1,yci1)*100))
plt.plot(sxci1.fit_transform(xci1),syci1.fit_transform(YPci1),'y',linewidth = 4,label = "Polynomial Regression : Accuracy = {}".format(reci1.score(Xci1,yci1)*100))
plt.plot(SXci1,SVYci1,'k',linewidth = 4,label ="SVR : Accuracy = {}".format(regci1.score(SXci1,SYci1)*100))
plt.title("Total Positive Cases: India")
plt.xlabel("Days")
plt.ylabel("Cases")
plt.legend()
plt.gcf().set_size_inches((10, 7))
```
